## Operators

Operators let us perform calculations and comparisons on values.

- Arithmetic: `+`, `-`, `*`, `/`, `//` (integer division), `%` (remainder)
- Comparison: `>`, `<`, `>=`, `<=`, `==`, `!=`
- Logical: `and`, `or`, `not`


```python
# Arithmetic Operations
a=10
b=3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Float Division:", a / b)
print("Integer Division:", a // b)
print("Remainder:", a % b)
```

```python
# Comparison Operators
x = 5
y = 10
print("x == y:", x == y)
print("x != y:", x != y)
print("x < y:", x < y)
print("x > y:", x > y)
print("x <= y:", x <= y)
print("x >= y:", x >= y)
```

```python
# Logical Operators
a = True   
b = False  
print("a and b:", a and b)
print("a or b:", a or b)    
print("not a:", not a)
print("not b:", not b)
```

### More if / elif / else examples

Here I practice different conditions: age categories, grading system, and finding the largest of three numbers.


```python
# Control Flow with if-elif-else
# Example: Age classification
age = 19

if age < 13:
    print("You are a child.")
elif age < 20:
    print("You are a teenager.")
elif age < 60:
    print("You are an adult.")
else:
    print("You are a senior.")
```

```python
# Example: Grading system
# Given a score, print the corresponding grade

marks = 82

if marks >= 90:
    print("Grade: A")
elif marks >= 80:
    print("Grade: B")
elif marks >= 70:
    print("Grade: C")
elif marks >= 60:
    print("Grade: D")
else:
    print("Grade: F")
```

```python
# Find the largest of three numbers
# Given three numbers, determine the largest

a = 12
b = 7
c = 15

if a >= b and a >= c:
    largest = a
elif b >= c:
    largest = b
else:
    largest = c

print("Largest number is:", largest)
```

```python
# Example 1: Check temperature
temperature = 32

if temperature >= 35:
    print("It is very hot.")
elif temperature >= 25:
    print("The weather is warm.")
else:
    print("It is cool or cold.")

# Example 2: Even or odd
number = 17

if number % 2 == 0:
    print(number, "is even")
else:
    print(number, "is odd")
```

## Loops in Python

Loops let us repeat a block of code multiple times.

- `for` loops: used when we know how many times we want to repeat (for each item in a list, each character in a string, etc.).
- `while` loops: used when we want to repeat until a condition becomes False.


```python
# For loop over a string
word = "datascience"

for ch in word:
    print("Letter:", ch)

# For loop over a list
cities = ["Kolkata", "Mumbai", "Delhi", "Bangalore"]

for city in cities:
    print("City:", city)
```

```python
# While loop: simulate guesses without input()

secret_number = 7
guesses = [3, 9, 7]  # pretend these are user guesses
index = 0

while index < len(guesses):
    guess = guesses[index]
    print("Trying guess:", guess)

    if guess < secret_number:
        print("Too low, try again.\n")
    elif guess > secret_number:
        print("Too high, try again.\n")
    else:
        print("Correct! You guessed it.")
        break  # stop the loop when correct

    index += 1
```

The `while` loop stops when the correct number is guessed or when all guesses are used.


## Functions in Python

Functions are reusable blocks of code.  
We define them once using `def` and then call them many times with different inputs.


### Example 1: greet(name)

This function takes a person's name as input and prints a greeting message.


```python
def greet(name):
    """Print a simple greeting message."""
    print("Hello,", name + "!")
    print("Welcome to my 30 days of data science journey.")

# Calling the function with different names
greet("Valor")          # change to your name if you want
greet("Data Scientist")
```

### Example 2: max_of_two(a, b)

This function compares two numbers and returns the larger one.


```python
def max_of_two(a, b):
    """Return the larger of two numbers."""
    if a >= b:
        return a
    else:
        return b

print("Max of 10 and 7 is:", max_of_two(10, 7))
print("Max of 3 and 9 is:", max_of_two(3, 9))
```
